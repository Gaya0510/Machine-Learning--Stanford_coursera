# Example1 Gradient descent simple
### $$ f(x) = x^2 +x+1$$

```python
x_1 = np.linspace(start=-3,stop=3, num=500)
```

```python
def f(x):
    return x**2 + x + 1
```

```python
def df(x):
    return 2*x + 1
```

```python
fig, ax = plt.subplots(1,1,figsize =(8,5))
ax.set_xlabel("X", fontsize=16)
ax.set_ylabel("f(x)",fontsize=16)
ax.set_xlim(-3,3)              
ax.set_ylim(0,8)
ax.plot(x_1, f(x_1), color="blue",lw =3, alpha = 0.8,)
ax.set_title("X vs f(x)", fontsize = 20)
```

```python
fig, ax = plt.subplots(1,2,figsize =(12,4))
ax[0].set_xlabel("X", fontsize=16)
ax[0].set_ylabel("f(x)",fontsize=16)
ax[0].set_xlim(-3,3)              
ax[0].set_ylim(0,8)
ax[0].plot(x_1, f(x_1), color="blue",lw =3, alpha = 0.8,)
ax[0].set_title("X vs f(x)", fontsize = 20)

ax[1].set_xlabel("X", fontsize=16)
ax[1].set_ylabel("df(x)",fontsize=16)
ax[1].set_xlim(-2,3)              
ax[1].set_ylim(-3,6)
ax[1].grid(True)
ax[1].plot(x_1, df(x_1), color="skyblue",lw =3, alpha = 0.8,)
ax[1].set_title("X vs df(x)", fontsize = 20)
```

```python
#gradient descent funtion 
def gradient_1(derivative_func, initial_guess_x, learning_rate=0.02, precision=0.001, loop_num=300):
    new_x = initial_guess_x
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    
    for i in range(loop_num):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - learning_rate * gradient
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        step_size = abs(new_x - previous_x)
        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
local_minima, x_list, slope_list = gradient_1(df, 3,0.1,0.00001,500)
print("Local minima is: ",local_minima)
print("Slope or df(x) for that point is: ", df(local_minima))
print("f(x) or cost value for that point is: ", f(local_minima))
```

```python
fig, ax = plt.subplots(1,2,figsize =(12,5))
ax[0].set_xlabel("X", fontsize=16)
ax[0].set_ylabel("f(x)",fontsize=16)
ax[0].set_xlim(-3,3)              
ax[0].set_ylim(0,8)
ax[0].plot(x_1, f(x_1), color="blue",lw =3, alpha = 0.8,)
ax[0].set_title("Cost Function", fontsize = 20)
ax[0].scatter(x_list, f(np.array(x_list)), color="red", s=50, alpha=0.5)

ax[1].set_xlabel("X", fontsize=16)
ax[1].set_ylabel("df(x)",fontsize=16)
ax[1].set_xlim(-2,3)              
ax[1].set_ylim(-3,6)
ax[1].grid(True)
ax[1].plot(x_1, df(x_1), color="skyblue",lw =3, alpha = 0.8,)
ax[1].set_title("Slope of Cost Function", fontsize = 20)
ax[1].scatter(x_list, np.array(slope_list), color="red", s=50, alpha=0.5)
```

```python
local_minima, x_list, slope_list = gradient_1(dg, 0.5, 0.02)
print("Local minima is: ",local_minima)
print("Slope or dg(x) for that point is: ", dg(local_minima))
print("g(x) or cost value for that point is: ", g(local_minima))
fig, ax = plt.subplots(1,2,figsize =(16,7))
ax[0].set_xlabel("X", fontsize=16)
ax[0].set_ylabel("g(x)",fontsize=16)
ax[0].set_xlim(-2,2)              
ax[0].set_ylim(0.5,5.5)
ax[0].plot(x_2, g(x_2), color="blue",lw =3, alpha = 0.8,)
ax[0].set_title("Cost Function", fontsize = 20)
ax[0].scatter(x_list, g(np.array(x_list)), color="red", s=50, alpha=0.5)

ax[1].set_xlabel("X", fontsize=16)
ax[1].set_ylabel("dg(x)",fontsize=16)
ax[1].set_xlim(-2,2)              
ax[1].set_ylim(-6,8)
ax[1].grid(True)
ax[1].plot(x_2, dg(x_2), color="skyblue",lw =3, alpha = 0.8,)
ax[1].set_title("Slope of Cost Function", fontsize = 20)
ax[1].scatter(x_list, np.array(slope_list), color="red", s=50, alpha=0.5)
```

# Example 3 - Divergence and overflow
### $$ f(x) = x^5 -2x^4+2$$

```python
x_3 = np.linspace(start=-2.5, stop=2.5, num=1000)
```

```python
def h(x):
    return x**5 - 2*x**4 +2

def dh(x):
    return 5*x**4 - 8*x**3 
```

```python
fig, ax = plt.subplots(1,2,figsize =(16,6))
ax[0].set_xlabel("X", fontsize=16)
ax[0].set_ylabel("h(x)",fontsize=16)
ax[0].set_xlim(-1.2,2.5)              
ax[0].set_ylim(-1,4)
ax[0].plot(x_3, h(x_3), color="blue",lw =3, alpha = 0.8,)
ax[0].set_title("X vs h(x)", fontsize = 20)

ax[1].set_xlabel("X", fontsize=16)
ax[1].set_ylabel("dh(x)",fontsize=16)
ax[1].set_xlim(-1,2)              
ax[1].set_ylim(-4,5)
ax[1].grid(True)
ax[1].plot(x_3, dh(x_3), color="skyblue",lw =3, alpha = 0.8,)
ax[1].set_title("X vs dh(x)", fontsize = 20)
```

```python
local_min, list_x, slope_list = gradient_1(derivative_func=dh, initial_guess_x= -0.2, 
                                                loop_num=71)
plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)

plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)', fontsize=16)

plt.plot(x_3, h(x_3), color='blue', linewidth=3, alpha=0.8)
plt.scatter(list_x, h(np.array(list_x)), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.grid()
plt.xlim(-1, 2)
plt.ylim(-4, 5)

plt.plot(x_3, dh(x_3), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(list_x, slope_list, color='red', s=100, alpha=0.5)

plt.show()

print('Local min occurs at: ', local_min)
print('Cost at this minimum is: ', h(local_min))
print('Number of steps: ', len(list_x))
```
